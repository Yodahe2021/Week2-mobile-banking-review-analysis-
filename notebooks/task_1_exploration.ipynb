```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_raw = pd.read_csv("../data/raw/reviews_raw.csv")
df_raw.head()
```

```python
df_raw.info()
df_raw.describe()
```

```python
df_raw['rating'].value_counts().plot(kind='bar')
plt.title('Rating Distribution (Raw Data)')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
df_raw['bank'].value_counts().plot(kind='bar')
plt.title('Review Count by Bank')
plt.show()
```
