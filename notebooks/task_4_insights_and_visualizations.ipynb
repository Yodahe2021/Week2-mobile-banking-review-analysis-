```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")

# IMPORTANT: Ensure this path is correct for your file
FILE_PATH = "../data/processed/reviews_task2_with_sentiment_and_themes.csv"
df = pd.read_csv(FILE_PATH)

# --- A. CALCULATE COMPETITIVE METRICS ---
# Group by bank and calculate the mean for Rating and Sentiment
comparison_df = df.groupby('bank')[['rating', 'sentiment_score']].mean().reset_index()

# Rename columns for clarity in the plot
comparison_df.columns = ['Bank', 'Average Rating', 'Average Sentiment']

# --- B. EXTRACT TOP THEMES FOR CBE (Based on your previous output) ---
# Mapping your calculated high-frequency keywords to actionable business themes
cbe_top_themes = {
    'Functional Instability (Doesn\'t Work)': 168 + 138 + 130, 
    'Update and Bug Issues': 194,
    'Other/Noise': 308 
}

# --- C. PRINT DATA FOR REPORT ---
print("--- Final Competitive Data ---")
print(comparison_df.to_markdown(index=False))

print("\n--- CBE Top Theme Counts (Aggregated) ---")
print(pd.Series(cbe_top_themes).sort_values(ascending=False).to_markdown())
```
